**Ejercicio 3: El Buscador de Nombres 游댍**

Este entrena el filtrado de texto.

- Lista: nombres = ["Antonio", "Maria", "Ana", "Pedro", "Alvaro"].

- Crea una funci칩n buscar_por_letra(lista, letra). (Recibe dos par치metros).

- La funci칩n debe devolver una lista nueva que solo contenga los nombres que empiezan por esa letra (usa nombre.startswith(letra)).

- Llama a la funci칩n buscando los nombres que empiecen por "A".

```python
nombres = ["Antonio", "Maria", "Ana", "Pedro", "Alvaro"]

def buscar_por_letra(lista,letra):
  lista_nueva=[]
  for i in lista:
    if i.startswith(letra):    #Simplemente utilizamos if i.startswith(letra):, nada de i.startwith(letra)==letra. Python entiende que si es True, debe entrar.
      lista_nueva.append(i)
  return lista_nueva
resultado=buscar_por_letra(nombres,"A")
print(resultado)
```

**쯏 si en vez de poner letra "A", pongo "a"?**

Para Python, "A" y "a" son caracteres totalmente distintos (tienen c칩digos internos diferentes). Si tu lista tiene "Antonio" y t칰 buscas con "a", el resultado ser치 una lista vac칤a [], porque "Antonio" empieza con "A" may칰scula.

쮺칩mo solucionarlo para que sea "inteligente"?
Los programadores usamos un truco: normalizamos ambos lados de la comparaci칩n a min칰sculas. As칤 no importa c칩mo lo escriba el usuario.

```python
def buscar_por_letra(lista, letra):
    lista_nueva = []
    letra_minuscula = letra.lower() # Convertimos la b칰squeda a min칰scula

    for i in lista:
        # Convertimos el nombre a min칰scula SOLO para la comparaci칩n
        if i.lower().startswith(letra_minuscula):
            lista_nueva.append(i)

    return lista_nueva
```
